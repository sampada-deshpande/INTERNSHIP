1ï¸âƒ£ if Statement

ðŸ‘‰ Executes code only if the condition is true.

âœ… Syntax:
if condition:
    statement(s)

EXAMPLE:

```python
age = 20

if age >= 18:
    print("You are eligible to vote.")
```

2ï¸âƒ£ if-else Statement

ðŸ‘‰ Chooses between two options.

âœ… Syntax:
if condition:
    statements_if_true
else:
    statements_if_false

EXAMPLE:

```python
num = 5

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

3ï¸âƒ£ if-elif-else Statement

ðŸ‘‰ Used when there are multiple conditions.

âœ… Syntax:
if condition1:
    block1
elif condition2:
    block2
elif condition3:
    block3
else:
    default_block

EXAMPLE:

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

4ï¸âƒ£ Nested if Statement

ðŸ‘‰ An if inside another if.
Used when a decision depends on another decision.

âœ… Syntax:
if condition1:
    if condition2:
        statements
        
EXAMPLE:

```python
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("You can enter.")
    else:
        print("ID required.")
else:
    print("Underage.")
```

```python
#GRADE TRACKER 
marks = int(input("Enter your marks (0-100): "))

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
elif marks >= 40:
    print("Grade: D")
else:
    print("Grade: F (Fail)")
```

```python
#ATM SYSTEM
balance = 10000

print("----- ATM MENU -----")
print("1. Check Balance")
print("2. Withdraw Money")
print("3. Deposit Money")

choice = int(input("Enter your choice (1-3): "))

if choice == 1:
    print("Your balance is:", balance)

elif choice == 2:
    withdraw = int(input("Enter amount to withdraw: "))
    if withdraw <= balance:
        balance = balance - withdraw
        print("Withdrawal successful")
        print("Remaining balance:", balance)
    else:
        print("Insufficient balance")

elif choice == 3:
    deposit = int(input("Enter amount to deposit: "))
    balance = balance + deposit
    print("Deposit successful")
    print("Updated balance:", balance)

else:
    print("Invalid choice")
```
