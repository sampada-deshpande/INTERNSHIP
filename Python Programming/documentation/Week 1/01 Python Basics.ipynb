ðŸ”¹ Virtualization

Virtualization is a technology that allows multiple operating systems or environments
to run on a single physical machine by sharing hardware resources efficiently.
It helps in better resource utilization, isolation, and cost reduction.

ðŸ”¹ Virtual Machine (VM)

A Virtual Machine is a software-based computer that runs inside a physical system.
Each VM has its own operating system and resources like CPU, RAM, and storage.
Examples of virtualization software:
1.VirtualBox
2.VMware

```python
# EMPLOYEE BONUS CALCULATOR

salary_input = input("Enter salary: ")

if salary_input.strip() == "":
    print("Salary cannot be empty")
else:
    salary = float(salary_input)
    years = int(input("Enter years of service: "))

    if years >= 5:
        bonus = salary * 0.10
    else:
        bonus = salary * 0.05

    print("Bonus Amount:", bonus)
    print("Total Salary:", salary + bonus)
```
