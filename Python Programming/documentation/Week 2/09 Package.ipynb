 PACKAGES IN PYTHON 

Python packages are a way to organize and structure code by grouping related modules into directories.

A package is essentially a folder that contains an __init__.py file and one or more Python files (modules).
Allows modules to be easily shared and distributed across different applications.

Key Components of a Python Package
1.Module: A single Python file containing reusable code (e.g., math.py).
2.Package: A directory containing modules and a special __init__.py file.
3.Sub-Packages: Packages nested within other packages for deeper organization.

Why Use Packages?

✔ Organize code
✔ Avoid name conflicts
✔ Reuse code
✔ Make projects scalable
✔ Easy collaboration

Types of Packages in Python

1️⃣ Built-in (Standard Library) Packages
Come with Python installation.

1.math – math functions
2.os – operating system tasks
3.sys – system-specific params
4.datetime – date & time
5.random – random values
6.json – JSON handling

EXAMPLE:
import math

print(math.sqrt(16))   # 4.0
print(math.pi)        # 3.141592653589793

2️⃣ Third-party Packages
Installed using pip.

1.Numpy – numerical computing
2.pandas – data analysis
3.matplotlib – plotting
4.requests – HTTP requests
5.flask / django – web apps


Install:

pip install numpy

Use:

import numpy as np

arr = np.array([1, 2, 3])
print(arr * 2)   # [2 4 6]
