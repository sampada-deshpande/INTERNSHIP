INHERITANCE IN PYTHON 

Inheritance is an Object-Oriented Programming (OOP) concept where one class (child/derived class) acquires the properties and methods of another class (parent/base class).

 It helps in code reusability, maintainability, and logical class structure.

Basic Syntax:
class Parent:
    # parent class code
    pass

class Child(Parent):
    # child class code
    pass


Example:

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()   # inherited method
d.bark()    # own method
```

1.Single Inheritance

Single Inheritance is a type of inheritance where one child (derived) class inherits from only one parent (base) class.

 This means:
1 Parent class
1 Child class

Syntax:

class Parent:
    # parent class code
    pass

class Child(Parent):
    # child class code
    pass


üîπ Real-World Example

Parent class: Vehicle

Child class: Car


```python
class Vehicle:
    def start(self):
        print("Vehicle starts")

class Car(Vehicle):
    def drive(self):
        print("Car is driving")
```

2.Multiple Inheritance
Multiple Inheritance is a type of inheritance where one child (derived) class inherits from more than one parent (base) class.

This means:
2 or more parent classes
1 child class

Syntax:

class Parent1:
    pass

class Parent2:
    pass

class Child(Parent1, Parent2):
    pass


Real World Example:
Smartphone

A smartphone is:

A Phone
A Camera
A Music Player


```python
class Father:
    def skill1(self):
        print("Driving")

class Mother:
    def skill2(self):
        print("Cooking")

class Child(Father, Mother):
    def skill3(self):
        print("Coding")

c = Child()
c.skill1()
c.skill2()
c.skill3()
```

The super()Method 
he super() method is used to call methods or the constructor of a parent (base) class from a child (derived) class.
 It is mainly used in inheritance to avoid rewriting parent class code.

üîπ Why Do We Use super()?

‚úî To reuse parent class code
‚úî To call parent class __init__()
‚úî To avoid hard-coding parent class names
‚úî To support multiple inheritance (MRO)

Syntax:
super().method_name()

Example:
super().__init__(arguments)


```python
#Multiple Inheritance With Super()Method
class A:
    def show(self):
        print("Class A")

class B:
    def show(self):
        print("Class B")

class C(A, B):
    def show(self):
        super().show()
        print("Class C")

obj = C()
obj.show()
```

3.Multilevel Inheritance
Multilevel Inheritance is a type of inheritance where a class inherits from another class, which itself inherits from another class.
for example:
Grandparent ‚Üí Parent ‚Üí Child

Syntax:
class A:
    pass

class B(A):
    pass

class C(B):
    pass

Real World Example:
Person ‚Üí Employee ‚Üí Manager



```python
class Animal:
    def eat(self):
        print("Animal eats")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Puppy(Dog):
    def play(self):
        print("Puppy plays")

p = Puppy()
p.eat()    # from Animal
p.bark()   # from Dog
p.play()   # from Puppy
```

```python
#Multilevel Inheritance with Super()method
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        super().show()
        print("B")

class C(B):
    def show(self):
        super().show()
        print("C")

obj = C()
obj.show()
```

4.Hierarchical Inheritance

Hierarchical Inheritance is a type of inheritance where multiple child classes inherit from the same single parent class.
    
In simple words:
One Parent ‚Üí Many Children

Syntax:

class Parent:
    pass

class Child1(Parent):
    pass

class Child2(Parent):
    pass

Real World Example:
Vehicle System

           Vehicle
          /   |    \
        Car  Bike  Bus


```python
class Vehicle:
    def start(self):
        print("Vehicle starts")

class Car(Vehicle):
    def drive(self):
        print("Car is driving")

class Bike(Vehicle):
    def ride(self):
        print("Bike is riding")

class Bus(Vehicle):
    def carry_passengers(self):
        print("Bus carries passengers")

c = Car()
b = Bike()
bus = Bus()

c.start()
c.drive()

b.start()
b.ride()

bus.start()
bus.carry_passengers()
```

Method Overiding in Python 

Method overriding happens when a child class provides its own implementation of a method that already exists in the parent class, using:
Same method name
Same parameters
Inheritance
The child class replaces the parent class method.
Example:

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):   # overridden method
        print("Dog barks")

d = Dog()
d.sound()
```

5.Hybrid Inheritance
Hybrid inheritance is a type of inheritance that is formed by combining two or more different types of inheritance
(such as multiple + multilevel, or hierarchical + multiple).

 In simple words:
Hybrid inheritance = mix of inheritance types

Why Hybrid Inheritance is Used?

‚úî To model complex real-world relationships
‚úî To reuse code efficiently
‚úî To organize classes logically

Python supports hybrid inheritance because it supports multiple inheritance.

Real World Scenario:

Consider a company where:
Every Person has basic details
An Employee is a Person
A Trainer is also a Person
A TechLead is both an Employee and a Trainer

           Person
          /      \
     Employee    Trainer
          \      /
           TechLead


```python
class Person:
    def details(self):
        print("Person details")

class Employee(Person):
    def work(self):
        print("Employee works")

class Trainer(Person):
    def teach(self):
        print("Trainer teaches")

class TechLead(Employee, Trainer):
    def lead(self):
        print("TechLead leads the team")

t = TechLead()
t.details()
t.work()
t.teach()
t.lead()
```
