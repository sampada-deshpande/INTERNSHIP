âœ… What is an Abstract Base Class (ABC) in Python?

An Abstract Base Class (ABC) is a class that cannot be instantiated and is used to define a blueprint (structure) for other classes.

ðŸ‘‰ It tells child classes what methods they must implement,
ðŸ‘‰ but it does not provide full implementation itself.

ðŸ”¹ Why do we need an Abstract Base Class?

âœ” To enforce a common structure
âœ” To achieve abstraction
âœ” To ensure consistency in large projects
âœ” To hide implementation details

Syntax of Abstract Base Class(abc):

from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

âœ” Shape is an abstract base class
âœ” area() is an abstract method

âœ…What is an Abstract Method in Python?

An abstract method is a method that is declared but not implemented in an abstract base class.

ðŸ‘‰ It only defines the method name,
ðŸ‘‰ It does not contain logic,
ðŸ‘‰ It must be implemented by child classe

Syntax of Abstract Method:

from abc import ABC, abstractmethod

class Parent(ABC):

    @abstractmethod
    def show(self):
        pass


âœ” show() is an abstract method
âœ” No implementation
âœ” Forces child class to define it



```python
#payment system

from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self):
        pass

class CardPayment(Payment):
    def pay(self):
        print("Payment done using Card")

class UPIPayment(Payment):
    def pay(self):
        print("Payment done using UPI")

p1 = CardPayment()
p2 = UPIPayment()

p1.pay()
p2.pay()
```

Diffrence Between the Data Abstraction And the Data Encapsulation

| Feature        | **Data Abstraction**                | **Data Encapsulation**             |
| -------------- | ----------------------------------- | ---------------------------------- |
| Meaning        | Hides **implementation details**    | Hides **data (variables)**         |
| Focus          | **What** an object does             | **How** data is protected          |
| Purpose        | Reduce complexity                   | Improve security                   |
| Achieved using | Abstract classes & abstract methods | Access modifiers & getters/setters |
| Python tools   | `ABC`, `@abstractmethod`            | `_protected`, `__private`          |
| User access    | Only essential methods              | Controlled data access             |
| Example        | ATM withdraw feature                | Bank account balance protection    |
